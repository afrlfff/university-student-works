```python
import matplotlib.pyplot as plt
import os
```

```python
path = os.path.join(os.getcwd(), "..", "..", "output", "results.txt")
```

```python
FILENAME = []
ENTROPY = []
START_SIZE = []
ENCODED_SIZE = []
ENCODING_TIME = []
DECODING_TIME = []
ENCODING_COEFFS = []

with open(path, "r", encoding="utf-8") as f:
    for line in f.readlines()[2:]:
        values = line.split()

        FILENAME.append(values[0])
        ENTROPY.append(float(values[1]))
        START_SIZE.append(float(values[2]))
        ENCODED_SIZE.append(float(values[3]))
        ENCODING_TIME.append(float(values[4]))
        DECODING_TIME.append(float(values[5]))
        ENCODING_COEFFS.append(float(values[2]) / float(values[3]))
```

```python
# check entropy
for i in range(len(FILENAME)):
    print(f"{FILENAME[i]} -> {ENTROPY[i]}")

# show entropy bar
plt.figure(figsize=(12, 6))
plt.bar(FILENAME, ENTROPY, color='blue')
plt.show()
```

```python
# temp
for i in range(len(FILENAME)):
    print(FILENAME[i], int(START_SIZE[i]), int(ENCODED_SIZE[i]), round(ENCODING_COEFFS[i], 3))
```

```python
# check compression
SORTED_ENTROPY, SORTED_ENCODING_COEFFS = zip(*sorted(zip(ENTROPY, ENCODING_COEFFS)))
plt.plot(SORTED_ENTROPY, SORTED_ENCODING_COEFFS, marker='o', color='blue')
plt.xlabel('Entropy', fontsize=12)
plt.ylabel('Compression ratio', fontsize=12)
plt.grid(True, color='black', linestyle='dashed', linewidth=0.75, alpha=0.75)
plt.show()
```
