```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calculate_Q(f0, fmin, fmax):
    return f0 / (fmax - fmin)
```

```python
f0 = 50
percents = [0, 30, 50, 70, 80, 90, 95, 98]

Q = [
    calculate_Q(f0, 15.027, 164.609),
    calculate_Q(f0, 19.862, 124.713),
    calculate_Q(f0, 24.840, 99.768),
    calculate_Q(f0, 32.114, 77.105),
    calculate_Q(f0, 37.003, 66.977),
    calculate_Q(f0, 42.848, 57.834),
    calculate_Q(f0, 46.167, 53.669),
    calculate_Q(f0, 48.282, 51.296),
]
```

```python
for i in range(len(Q)):
    print(f"{percents[i]}% -> Q = {Q[i]:.3f}") 
```

```python
plt.title("")
plt.xlabel("%")
plt.ylabel("Q")
plt.plot(percents, Q, color='blue', marker='o')
plt.grid(True)
plt.xticks(np.arange(0, 100+10, 10))
plt.yticks(np.arange(0, 18+1, 1))
plt.show()
```
